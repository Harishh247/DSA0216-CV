```python
# Step 1: Install OpenCV
!pip install opencv-python

# Step 2: Import required libraries
import cv2
import numpy as np
from google.colab import files
from matplotlib import pyplot as plt

# Step 3: Upload image
uploaded = files.upload()

# Get uploaded file name
image_path = list(uploaded.keys())[0]

# Step 4: Read image
image = cv2.imread(image_path)
original = image.copy()

# Convert BGR to RGB for displaying
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 5: Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 6: Apply Gaussian Blur (to remove noise)
blur = cv2.GaussianBlur(gray, (5,5), 0)

# Step 7: Canny Edge Detection
edges = cv2.Canny(blur, 100, 200)

# Step 8: Find Contours
contours, hierarchy = cv2.findContours(edges,
                                       cv2.RETR_EXTERNAL,
                                       cv2.CHAIN_APPROX_SIMPLE)

# Step 9: Draw Contours
cv2.drawContours(original, contours, -1, (0,255,0), 2)

# Convert BGR to RGB for display
output_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)

# Step 10: Display Results
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edges")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(output_rgb)
plt.title("Contour Detection")
plt.axis("off")

plt.show()

print("Total Contours Detected:", len(contours))
```
